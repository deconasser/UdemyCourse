```python
# A Corner is a point whose local neighborhood stands in two dominant and different edges direction
# A corner can be interpreted as a junction of two edges where an edge is a sudden change in image brightness
# 2 popular algorithms:
# + Harris Corner Detection (1988) R = lamda1*lamda2 - k(lamda1+ lamda2)
# + Shi-Tomashi Corner Detection (1994) R = min(lamda1, lamda2)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/flat_chessboard.png', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
real_chess = cv2.imread("/content/real_chessboard.jpg")
res_real_chess = real_chess.copy()
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(real_chess, cmap ="gray")
```

```python
img = np.float32(img)
real_chess = np.float32(real_chess)
```

```python
dst = cv2.cornerHarris(img, 2, 3, 0.04)
plt.imshow(dst, cmap='gray')
```

```python
dst = cv2.cornerHarris(real_chess, 2, 3, 0.04)
dst = cv2.dilate(dst, None)
res_real_chess[dst>0.01*dst.max()] = [255, 0, 0]
plt.imshow(res_real_chess)
```
