```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('/content/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
face = cv2.imread("/content/sammy_face.jpg")
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
  fully_copy = full.copy()
  method = eval(m)
  result = cv2.matchTemplate(fully_copy, face, method)
  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
  if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
    location = min_loc
  else:
    location = max_loc
  cv2.rectangle(fully_copy, location, (location[0] + face.shape[1], location[1] + face.shape[0]), (255, 0, 0), 10)
  plt.subplot(121),
  plt.imshow(result)
  plt.title('Matching Result'),
  plt.subplot(122),
  plt.imshow(fully_copy)

  plt.show()
  print('\n')
```
