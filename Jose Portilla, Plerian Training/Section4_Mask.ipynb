```python
# co 2 cach de chen mask 1. addWeight 2. bitwise_or
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_1 = cv2.imread("/content/dog_backpack.png")
img_2 = cv2.imread("/content/watermark_no_copy.png")
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
```

```python
print(img_1.shape)
print(img_2.shape)
```

```python
img_1 = cv2.resize(img_1, (1200, 1200))
img_2 = cv2.resize(img_2, (1200, 1200))
```

```python
fig = plt.figure(figsize=(4,6))
ax = fig.add_subplot(121)
plt.imshow(img_1)
ax = fig.add_subplot(122)
plt.imshow(img_2)
```

```python
blended = cv2.addWeighted(src1 = img_1, alpha = 0.8, src2 = img_2, beta=0.2, gamma=0)
plt.imshow(blended)
```

```python
img_1 = cv2.imread("/content/dog_backpack.png")
img_2 = cv2.imread("/content/watermark_no_copy.png")
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_2 = cv2.resize(img_2, (600, 600))
```

```python
fig = plt.figure(figsize=(4,6))
ax = fig.add_subplot(121)
plt.imshow(img_1)
ax = fig.add_subplot(122)
plt.imshow(img_2)
```

```python
x_offset = 1401-600
y_offset = 934 - 600
```

```python
crop_img = img_1[x_offset:1401, y_offset:934]
plt.imshow(crop_img)
```

```python
img_2 = cv2.cvtColor(img_2, cv2.COLOR_RGB2GRAY)
plt.imshow(img_2, cmap="gray")
```

```python
img_2 = cv2.bitwise_not(img_2)
img_2_arr = np.asarray(img_2)
print(img_2_arr)
plt.imshow(img_2, cmap = "gray")
```

```python
print(img_2.shape)
bg = np.full((600, 600, 3), 255, dtype=np.uint8)
bk = cv2.bitwise_or(bg, bg, mask=img_2)
res_img = cv2.bitwise_or(crop_img, bk)

```
