```python
img = cv2.imread("/content/gorilla.jpg", 0)
```

```python
def display_img(img):
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap="gray")
```

```python
display_img(img)
```

```python
hist = cv2.calcHist([img], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist)
```

```python
img_equalize = cv2.equalizeHist(img)
display_img(img_equalize)
```

```python
hist = cv2.calcHist([img_equalize], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist)
```
