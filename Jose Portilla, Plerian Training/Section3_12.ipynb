```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("/content/Starbucks Dog.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.imread("/content/Starbucks Dog.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
resize_img = cv2.resize(img, (1000, 400))
plt.imshow(resize_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(img, (0, 0), img, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
flip_img = cv2.flip(img, -1)
plt.imshow(flip_img)
```

```python
cv2.imwrite("/content/flip_img.jpg", flip_img)
```

```python
fig = plt.figure(figsize=(2,4))
ax = fig.add_subplot(121)
ax.imshow(img)
ax = fig.add_subplot(122)
ax.imshow(flip_img)
```
