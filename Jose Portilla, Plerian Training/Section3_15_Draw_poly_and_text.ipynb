```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
blank_img = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img, text="Hello World", org=(10, 500), fontFace=font, fontScale=2, color=(255,255,255), thickness= 3, lineType=cv2.LINE_4)
plt.imshow(blank_img)
```

```python
pol = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
pol = pol.reshape(-1, 1, 2)
cv2.polylines(blank_img, [pol], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```
