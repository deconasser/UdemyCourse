```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(384,0), pt2=(410,150), color=(0, 255, 0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(300, 200), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(100, 450), pt2=(300, 450), color=(0,0,255), thickness=3)
plt.imshow(blank_img)
```
