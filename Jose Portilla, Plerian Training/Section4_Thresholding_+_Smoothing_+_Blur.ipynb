```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('/content/rainbow.jpg', 0)
plt.imshow(image, cmap="gray")
```

```python
ret, thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap="gray")
```

```python
ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap="gray")
```

```python
img = cv2.imread("/content/crossword.jpg", 0)
plt.imshow(img, cmap='gray')
```

```python
adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                      cv2.THRESH_BINARY, 11, 8)
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111)
ax.imshow(adaptive_mean, cmap='gray')
```

```python
image = cv2.imread('/content/bricks.jpg', 0)
smoothing_img = cv2.GaussianBlur(image, (3, 3), 0)
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(121)
ax.imshow(image, cmap='gray')
ax = fig.add_subplot(122)
ax.imshow(smoothing_img, cmap='gray')
```
