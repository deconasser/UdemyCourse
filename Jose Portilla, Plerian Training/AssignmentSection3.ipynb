```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/49er decal.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
flip_img = cv2.flip(img, -1).copy()
plt.imshow(flip_img)
```

```python
rangle_img = cv2.rectangle(img, (250, 150), (600, 550), (255,0,0),thickness=4)
plt.imshow(rangle_img)
```

```python
blank_img = np.zeros((512, 512, 3), dtype=np.int16)
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img, [pts], True, (0,255,255))
plt.imshow(img)
```
